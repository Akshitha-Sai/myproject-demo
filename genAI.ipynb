```python
import pandas as pd
import gradio as gr

# Load the dataset
df = pd.read_csv("telugu_dataset.csv")

# Function to get translation
def translate(sentence):
    sentence = sentence.lower().strip()

    # Check for the sentence in English column
    row = df[df["English"].str.lower().str.strip() == sentence]

    if not row.empty:
        phonetics = row["Telugu Phonetics"].values[0]
        script = row["Telugu Script"].values[0]
        return phonetics, script
    else:
        return "Not found", "Please enter a valid sentence from the dataset"

# Create Gradio Interface
iface = gr.Interface(
    fn=translate,
    inputs=gr.Textbox(label="Enter an English sentence"),
    outputs=[gr.Textbox(label="Telugu Phonetics"), gr.Textbox(label="Telugu Script")],
    title="English to Telugu Translator",
    description="Enter an English sentence from the dataset to get its Telugu phonetics and script."
)

# Launch with debugging enabled
iface.launch(share=True, debug=True)
```
